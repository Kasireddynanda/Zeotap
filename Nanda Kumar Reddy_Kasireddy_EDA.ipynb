```python
# Relationships and Heatmaps
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
categorical_columns = ['CustomerID','CustomerName', 'Region','SignupDate']
Customers[categorical_columns] = Customers[categorical_columns].apply(label_encoder.fit_transform)
print(Customers)
```

```python
correlation_matrix = Customers.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
categorical_columns = ['ProductID', 'ProductName', 'Category']
Products[categorical_columns] = Products[categorical_columns].apply(label_encoder.fit_transform)
print(Products)
```

```python
correlation_pmatrix = Products.corr()
print(correlation_pmatrix)
```

```python
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(correlation_pmatrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
categorical_columns = ['TransactionID','CustomerID','ProductID','TransactionDate']
Transactions[categorical_columns] = Transactions[categorical_columns].apply(label_encoder.fit_transform)
print(Transactions)
```

```python
correlation_tmatrix = Transactions.corr()
print(correlation_tmatrix)
```

```python
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(correlation_tmatrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
